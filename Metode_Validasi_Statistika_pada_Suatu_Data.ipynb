**JARQUE BERA**

Merupakan pengujian statistika untuk mengetahui apakah data berdistribusi normal atau tidak.
Uji JB digunakan untuk menggantikan penilaian subjektif dari visualisasi grafis (seperti Q-Q Plot atau Histogram). Metode ini mengukur seberapa jauh karakteristik skewness (kemiringan) dan kurtosis (keruncingan) dari data sampel menyimpang dari karakteristik distribusi normal ideal.
Data dikatakan berdistribusi normal ideal jika nilai skewness (kemiringan) adalah 0 dan nilai kurtosis (keruncingan) adalah 3.

**RUMUS**
$$ JB = \frac{n}{6} \left( S^2 + \frac{(K-3)^2}{4} \right) $$

Keterangan:
- $JB$ = Statistik Jarque-Bera
- $n$ = Jumlah observasi (jumlah data)
- $S$ = Skewness (kemiringan)
- $K$ = Kurtosis (keruncingan)

Skewness ($S$) dihitung sebagai:
$$ S = \frac{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^3}{\left( \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2 \right)^{3/2}} $$

Kurtosis ($K$) dihitung sebagai:
$$ K = \frac{\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^4}{\left( \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2 \right)^{2}} $$

```python
def jarque_bera_test(data):

    # Step 1: Menghitung jumlah data (n)
    n = len(data)

    # Step 2: Menghitung nilai rata-rata (mean)
    mean = sum(data) / n

    # Step 3: Menghitung momen kedua (m2)
    # m2 = Σ(x-mean)^2 / n
    m2 = sum((x - mean)**2 for x in data) / n

    # Step 4: Menghitung momen ketiga (m3)
    # m3 = Σ(x-mean)^3 / n
    m3 = sum((x - mean)**3 for x in data) / n

    # Step 5: Menghitung momen keempat (m4)
    # m4 = Σ(x-mean)^4 / n
    m4 = sum((x - mean)**4 for x in data) / n

    # Step 6: Menghitung nilai Skewness
    # Skewness = m3 / (m2^(3/2))
    skewness = m3 / (m2 ** 1.5)

    # Step 7: Menghitung nilai Kurtosis
    # Kurtosis = m4 / (m2^2)
    kurtosis = m4 / (m2 ** 2)

    # Step 8: Menghitung statistik Jarque-Bera
    # JB = (n/6) * (Skewness^2 + ((Kurtosis-3)^2)/4)
    jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

    # Step 9: Menentukan nilai kritis Chi-Square
    # α = 0,05 dan df = 2
    chi_square_critical = 5.991

    # Step 10: Mengambil keputusan hipotesis
    if jb > chi_square_critical:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    else:
        keputusan = "Gagal menolak H0: data berdistribusi normal"

    # Step 11: Menampilkan seluruh hasil perhitungan
    print("n =", n)
    print("mean =", mean)
    print("m2 =", m2)
    print("m3 =", m3)
    print("m4 =", m4)
    print("skewness =", skewness)
    print("kurtosis =", kurtosis)
    print("jb =", jb)
    print("keputusan =", keputusan)

# Data yang diuji
data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]

# Menjalankan fungsi Jarque-Bera
jarque_bera_test(data)
```

**KOLMOGOROV SMIRNOV**

Metode statistik yang digunakan untuk mengetahui apakah data yang diteliti mengikuti distribusi normal atau tidak. Pengujian ini dilakukan dengan membandingkan distribusi data sampel dengan distribusi normal teoritis.
Hasil pengujian dilihat dari nilai signifikansi (p-value), di mana data dianggap berdistribusi normal jika nilai signifikansi lebih besar dari 0,05, dan dianggap tidak berdistribusi normal jika nilai signifikansi lebih kecil dari 0,05.


**RUMUS**
1. Mengihitung Rata Rata
$$
\bar{x} = \frac{\sum x_i}{n}
$$
    Keterangan:
    - $\bar{x}$ = rata-rata data
    - $x_i$ = data ke-i
    - $n$ = jumlah data
2. Standar Devisiasi
$$
s = \sqrt{\frac{\sum (x_i-\bar{x})^2}{n-1}}
$$
   Keterangan:
    - $s$ = standar deviasi
    - $x_i$ = data ke-i
    - $\bar{x}$ = rata-rata data
    - $n$ = jumlah data
3. Nilai Z
$$
Z_i = \frac{x_i-\bar{x}}{s}
$$
    Keterangan:
    - $Z_i$ = nilai Z data ke-i
    - $x_i$ = data ke-i
    - $\bar{x}$ = rata-rata data
    - $s$ = standar deviasi
4. Distribusi Empiris
$$
F_n(x)=\frac{i}{n}
$$
    Keterangan:
    - $F_n(x)$ = distribusi empiris
    - $i$ = urutan data setelah diurutkan
    - $n$ = jumlah data
5. Statistik Uji KS
$$
D = \max |F_n(x)-F(x)|
$$
    Keterangan:
    - $D$ = statistik Kolmogorov-Smirnov
    - $F_n(x)$ = distribusi empiris
    - $F(x)$ = distribusi normal teoritis
6. Nilai Kritis
$$
D_{tabel}=\frac{1.36}{\sqrt{n}}
$$
    Keterangan:
    - $D_{tabel}$ = nilai kritis Kolmogorov-Smirnov
    - $n$ = jumlah data

```python
import math

def kolmogorov_smirnov_test(data):

    # Step 1: Mengurutkan data dari kecil ke besar
    data.sort()

    # Step 2: Menghitung jumlah data (n)
    n = len(data)

    # Step 3: Menghitung rata-rata (mean)
    mean = sum(data) / n

    # Step 4: Menghitung varians
    varians = sum((x - mean)**2 for x in data) / n

    # Step 5: Menghitung standar deviasi
    std = varians ** 0.5

    # Variabel untuk menyimpan nilai D maksimum
    d_max = 0

    # Step 6: Menghitung statistik KS untuk setiap data
    for i in range(n):

        # Menghitung nilai Z
        z = (data[i] - mean) / std

        # Menghitung CDF distribusi normal teoritis
        ft = 0.5 * (1 + math.erf(z / math.sqrt(2)))

        # Menghitung distribusi kumulatif empiris
        fo = (i + 1) / n

        # Menghitung selisih absolut
        d = abs(fo - ft)

        # Menyimpan nilai D terbesar
        if d > d_max:
            d_max = d

    # Step 7: Menentukan nilai kritis KS
    d_kritis = 1.36 / (n ** 0.5)

    # Step 8: Mengambil keputusan
    if d_max > d_kritis:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    else:
        keputusan = "Gagal menolak H0: data berdistribusi normal"

    # Step 9: Menampilkan hasil
    print("Data terurut =", data)
    print("n =", n)
    print("mean =", mean)
    print("varians =", varians)
    print("standar deviasi =", std)
    print("D hitung =", d_max)
    print("D kritis =", d_kritis)
    print("Keputusan =", keputusan)

# Data yang diuji
data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]

# Menjalankan fungsi
kolmogorov_smirnov_test(data)
```

**UJI HOMOGENITAS VARIANS (F VARIANS)**

Metode statistik yang digunakan untuk mengetahui apakah dua kelompok data memiliki varians yang sama atau berbeda secara signifikan.



**RUMUS**

Untuk menguji homogenitas varians dari dua kelompok data, kita menggunakan statistik uji F. Langkah-langkahnya adalah sebagai berikut:

1.  **Hitung Varians Masing-masing Kelompok**
    Varians sampel untuk kelompok 1 ($s_1^2$) dan kelompok 2 ($s_2^2$) dihitung sebagai:

    $$ s_1^2 = \frac{\sum (x_{1i} - \bar{x}_1)^2}{n_1 - 1} $$
    $$ s_2^2 = \frac{\sum (x_{2i} - \bar{x}_2)^2}{n_2 - 1} $$

    Keterangan:
    - $x_{1i}$, $x_{2i}$ = data observasi pada kelompok 1 dan 2
    - $\bar{x}_1$, $\bar{x}_2$ = rata-rata kelompok 1 dan 2
    - $n_1$, $n_2$ = jumlah observasi pada kelompok 1 dan 2

2.  **Hitung Statistik Uji F**
    Statistik F dihitung dengan membagi varians yang lebih besar dengan varians yang lebih kecil:
    $$ F_{hitung} = \frac{s_{besar}^2}{s_{kecil}^2} $$
    Dimana $s_{besar}^2$ adalah varians yang lebih besar dan $s_{kecil}^2$ adalah varians yang lebih kecil dari kedua kelompok.

3.  **Tentukan Derajat Kebebasan (degrees of freedom)**
    - $df_1$ (pembilang) = $n_{besar} - 1$
    - $df_2$ (penyebut) = $n_{kecil} - 1$
    (dimana $n_{besar}$ adalah jumlah observasi dari kelompok dengan varians yang lebih besar, dan $n_{kecil}$ adalah jumlah observasi dari kelompok dengan varians yang lebih kecil).

4.  **Bandingkan dengan Nilai F Tabel**
    Nilai $F_{hitung}$ dibandingkan dengan nilai $F_{tabel}$ pada tingkat signifikansi ($\alpha$) tertentu dan derajat kebebasan $df_1$ dan $df_2$.

    - Jika $F_{hitung} > F_{tabel}$, maka tolak $H_0$ (varians tidak homogen).
    - Jika $F_{hitung} \le F_{tabel}$, maka gagal menolak $H_0$ (varians homogen).

```python
def uji_F_varians(data1, data2, alpha=0.05):

    # Step 1: Menghitung jumlah data masing-masing kelompok
    n1 = len(data1)
    n2 = len(data2)

    # Step 2: Menghitung rata-rata masing-masing kelompok
    mean1 = sum(data1) / n1
    mean2 = sum(data2) / n2

    # Step 3: Menghitung varians sampel data1
    varians1 = sum((x - mean1)**2 for x in data1) / (n1 - 1)

    # Step 4: Menghitung varians sampel data2
    varians2 = sum((x - mean2)**2 for x in data2) / (n2 - 1)

    # Step 5: Menentukan varians terbesar dan terkecil
    if varians1 > varians2:
        f_hitung = varians1 / varians2
    else:
        f_hitung = varians2 / varians1

    # Step 6: Menentukan nilai F tabel
    # α = 0.05, df1 = 4, df2 = 4
    f_tabel = 6.39

    # Step 7: Mengambil keputusan
    if f_hitung > f_tabel:
        keputusan = "Tolak H0: varians tidak homogen"
    else:
        keputusan = "Gagal menolak H0: varians homogen"

    # Step 8: Menampilkan hasil
    print("n1 =", n1)
    print("n2 =", n2)
    print("mean1 =", mean1)
    print("mean2 =", mean2)
    print("varians1 =", varians1)
    print("varians2 =", varians2)
    print("f_hitung =", f_hitung)
    print("f_tabel =", f_tabel)
    print("keputusan =", keputusan)

data1 = [70, 75, 80, 72, 78]
data2 = [65, 85, 60, 90, 70]

uji_F_varians(data1, data2)
```

**UJI LEVANE**

Pengujian statistik yang digunakan untuk mengetahui apakah varians atau keragaman data dari dua atau lebih kelompok memiliki nilai yang sama (homogen). Uji ini penting dilakukan sebagai salah satu syarat dalam analisis statistik parametrik, seperti uji t dan ANOVA.
Pada uji Levane, hipotesis nol (H₀) menyatakan bahwa varians antar kelompok adalah sama atau homogen, sedangkan hipotesis alternatif (H₁) menyatakan bahwa varians antar kelompok berbeda atau tidak homogen. Data dinyatakan homogen apabila nilai signifikansi (p-value) lebih besar dari 0,05, dan dinyatakan tidak homogen apabila nilai signifikansi lebih kecil dari 0,05.

**RUMUS**
1. Menghitung Median Tiap Kelompok
$$
\tilde{X}_i = \text{Median}(X_i)
$$
    Keterangan:
    - $\tilde{X}_i$ = median kelompok ke-i
    - $X_i$ = data pada kelompok ke-i
2. Menghitung Deviasi Absolut
$$
Z_{ij}=|X_{ij}-\tilde{X}_i|
$$
    Keterangan:
    - $Z_{ij}$ = deviasi absolut
    - $X_{ij}$ = data ke-j pada kelompok ke-i
    - $\tilde{X}_i$ = median kelompok ke-i
3. Rata-Rata Deviasi Absolut Tiap Kelompok
$$
\bar{Z}_i=\frac{\sum Z_{ij}}{n_i}
$$
    Keterangan:
    - $\bar{Z}_i$ = rata-rata deviasi absolut kelompok ke-$i$
    - $Z_{ij}$ = deviasi absolut data ke-$j$ pada kelompok ke-$i$
    - $n_i$ = jumlah data pada kelompok ke-$i$
4. Rata-Rata Deviasi Absolut Keseluruhan
$$
\bar{Z}=\frac{\sum Z_{ij}}{N}
$$
    Keterangan:
    - $\bar{Z}$ = rata-rata deviasi absolut seluruh kelompok
    - $Z_{ij}$ = deviasi absolut data ke-$j$ pada kelompok ke-$i$
    - $N$ = jumlah seluruh data
5. Menghitung Statistik Uji Levene
$$
W=
\frac{(N-k)}
{(k-1)}
\times
\frac{
\sum n_i(\bar Z_i-\bar Z)^2
}
{
\sum\sum(Z_{ij}-\bar Z_i)^2
}
$$
    Keterangan:
    - $W$ = statistik uji Levene
    - $N$ = jumlah seluruh data
    - $k$ = jumlah kelompok
    - $n_i$ = jumlah data kelompok ke-i
    - $\bar{Z}_i$ = rata-rata deviasi absolut kelompok ke-i
    - $\bar{Z}$ = rata-rata deviasi absolut seluruh kelompok

**SOAL**

```python
def uji_levene(data1, data2):

    # Step 1: Menghitung jumlah data tiap kelompok
    n1 = len(data1)
    n2 = len(data2)

    # Step 2: Menghitung total data
    N = n1 + n2

    # Step 3: Menentukan jumlah kelompok
    k = 2

    # Step 4: Menghitung rata-rata masing-masing kelompok
    mean1 = sum(data1) / n1
    mean2 = sum(data2) / n2

    # Step 5: Menghitung deviasi absolut terhadap rata-rata
    z1 = [abs(x - mean1) for x in data1]
    z2 = [abs(x - mean2) for x in data2]

    # Step 6: Menghitung rata-rata deviasi absolut tiap kelompok
    zbar1 = sum(z1) / n1
    zbar2 = sum(z2) / n2

    # Step 7: Menghitung rata-rata seluruh deviasi absolut
    zbar = (sum(z1) + sum(z2)) / N

    # Step 8: Menghitung pembilang statistik Levene
    pembilang = n1 * (zbar1 - zbar)**2 + n2 * (zbar2 - zbar)**2

    # Step 9: Menghitung penyebut statistik Levene
    penyebut = sum((z - zbar1)**2 for z in z1) + sum((z - zbar2)**2 for z in z2)

    # Step 10: Menghitung nilai statistik Levene (W hitung)
    w_hitung = ((N - k) / (k - 1)) * (pembilang / penyebut)

    # Step 11: Menentukan nilai F tabel
    # α = 0.05, df1 = k-1 = 1, df2 = N-k = 8
    f_tabel = 5.32

    # Step 12: Mengambil keputusan
    if w_hitung > f_tabel:
        keputusan = "Tolak H0: varians tidak homogen"
    else:
        keputusan = "Gagal menolak H0: varians homogen"

    # Step 13: Menampilkan seluruh hasil
    print("n1 =", n1)
    print("n2 =", n2)
    print("mean1 =", mean1)
    print("mean2 =", mean2)
    print("z1 =", z1)
    print("z2 =", z2)
    print("zbar1 =", zbar1)
    print("zbar2 =", zbar2)
    print("zbar =", zbar)
    print("w_hitung =", w_hitung)
    print("f_tabel =", f_tabel)
    print("keputusan =", keputusan)

data1 = [70, 75, 80, 72, 78]
data2 = [65, 85, 60, 90, 70]

uji_levene(data1, data2)
```

**UJI INDEPENDENSI CHI-SQUARE**

Metode statistik yang digunakan untuk mengetahui apakah terdapat hubungan antara dua variabel kategorik dalam suatu populasi. Uji ini tidak mengukur kuat atau lemahnya hubungan, melainkan hanya memeriksa apakah kedua variabel saling berhubungan atau bersifat independen.


**RUMUS**

Statistik uji Chi-Square dihitung dengan rumus:
$$ \chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{k} \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $$

Keterangan:
- $\chi^2$ = Statistik Chi-Square
- $O_{ij}$ = Frekuensi observasi (aktual) pada baris ke-$i$ dan kolom ke-$j$
- $E_{ij}$ = Frekuensi harapan (expected) pada baris ke-$i$ dan kolom ke-$j$
- $r$ = Jumlah baris
- $k$ = Jumlah kolom

Frekuensi harapan ($E_{ij}$) dihitung dengan rumus:
$$ E_{ij} = \frac{(Total Baris_i) \times (Total Kolom_j)}{Total Keseluruhan} $$

Derajat Kebebasan (degrees of freedom, $df$) dihitung sebagai:
$$ df = (r-1)(k-1) $$

**Kriteria Keputusan:**

1.  **Pendekatan Nilai Kritis:** Bandingkan $\chi^2_{hitung}$ dengan $\chi^2_{tabel}$ pada tingkat signifikansi ($\alpha$) dan $df$ yang sesuai.
    - Jika $\chi^2_{hitung} > \chi^2_{tabel}$, maka tolak $H_0$ (ada hubungan).
    - Jika $\chi^2_{hitung} \le \chi^2_{tabel}$, maka gagal menolak $H_0$ (tidak ada hubungan).

2.  **Pendekatan p-value:** Bandingkan p-value dengan tingkat signifikansi ($\alpha$).
    - Jika p-value < $\alpha$, maka tolak $H_0$ (ada hubungan).
    - Jika p-value $\ge \alpha$, maka gagal menolak $H_0$ (tidak ada hubungan).

**SOAL**

```python
from scipy.stats import chi2

def uji_chi_square(data, alpha=0.05):

    # Step 1: Menghitung jumlah baris dan kolom
    jumlah_baris = len(data)
    jumlah_kolom = len(data[0])

    # Step 2: Menghitung total baris
    total_baris = [sum(baris) for baris in data]

    # Step 3: Menghitung total kolom
    total_kolom = []
    for j in range(jumlah_kolom):
        total_kolom.append(sum(data[i][j] for i in range(jumlah_baris)))

    # Step 4: Menghitung grand total
    grand_total = sum(total_baris)

    # Step 5: Menghitung frekuensi harapan
    harapan = []

    for i in range(jumlah_baris):
        row = []
        for j in range(jumlah_kolom):
            eij = (total_baris[i] * total_kolom[j]) / grand_total
            row.append(eij)
        harapan.append(row)

    # Step 6: Menghitung Chi-Square hitung
    chi_square = 0

    for i in range(jumlah_baris):
        for j in range(jumlah_kolom):
            chi_square += ((data[i][j] - harapan[i][j])**2) / harapan[i][j]

    # Step 7: Menghitung derajat bebas
    df = (jumlah_baris - 1) * (jumlah_kolom - 1)

    # Step 8: Menghitung Chi-Square tabel
    chi_tabel = chi2.ppf(1 - alpha, df)

    # Step 9: Mengambil keputusan
    if chi_square > chi_tabel:
        keputusan = "Tolak H0: terdapat hubungan"
    else:
        keputusan = "Gagal menolak H0: tidak terdapat hubungan"

    # Step 10: Menampilkan hasil
    print("jumlah_baris =", jumlah_baris)
    print("jumlah_kolom =", jumlah_kolom)
    print("grand_total =", grand_total)
    print("harapan =", harapan)
    print("chi_square =", chi_square)
    print("chi_tabel =", chi_tabel)
    print("keputusan =", keputusan)

data = [
    [30, 20],
    [15, 35]]
uji_chi_square(data)
```

**INDEPENDENSI FISHER'S EXACT TEST**

Mtode statistik yang digunakan untuk mengetahui ada atau tidaknya hubungan antara dua variabel kategorik dalam sebuah tabel kontingensi, terutama ketika jumlah sampel relatif kecil atau terdapat frekuensi harapan kurang dari 5 pada beberapa sel.
Uji ini merupakan alternatif dari uji Chi-Square yang memberikan hasil lebih akurat untuk data berukuran kecil.
Pada Fisher's Exact Test, hipotesis nol (H₀) menyatakan bahwa tidak terdapat hubungan atau kedua variabel bersifat independen, sedangkan hipotesis alternatif (H₁) menyatakan bahwa terdapat hubungan antara kedua variabel. Keputusan pengujian didasarkan pada nilai signifikansi (p-value), di mana H₀ diterima jika nilai p-value > 0,05 dan H₀ ditolak jika nilai p-value < 0,05.

**RUMUS-RUMUS**
1. Bentuk Tabel Kontingensi
Misalkan tabel kontingensi 2×2 sebagai berikut:

    |      | Kolom 1 | Kolom 2 |
    |------|---------|---------|
    | Baris 1 | a | b |
    | Baris 2 | c | d |

    Keterangan:
    - $a$ = frekuensi pada baris 1 kolom 1
    - $b$ = frekuensi pada baris 1 kolom 2
    - $c$ = frekuensi pada baris 2 kolom 1
    - $d$ = frekuensi pada baris 2 kolom 2
2. Menghitung Jumlah Seluruh Data
$$
n = a+b+c+d
$$
    Keterangan:
    - $n$ = jumlah seluruh data
    - $a,b,c,d$ = frekuensi pada masing-masing sel
3. Menghitung Faktorial
$$
n! = n \times (n-1) \times (n-2) \times \cdots \times 1
$$
    Keterangan:
    - $n!$ = nilai faktorial dari n
    - $n$ = bilangan bulat positif

4. Rumus Probabilitas Fisher
$$
P =\frac{(a+b)!(c+d)!(a+c)!(b+d)!}
{
a!b!c!d!n!
}
$$
    Keterangan:
    - $P$ = probabilitas Fisher's Exact Test
    - $a$ = frekuensi pada baris 1 kolom 1
    - $b$ = frekuensi pada baris 1 kolom 2
    - $c$ = frekuensi pada baris 2 kolom 1
    - $d$ = frekuensi pada baris 2 kolom 2
    - $n$ = jumlah seluruh data
    - $!$ = operasi faktorial

```python
from math import factorial

def fisher_exact_test(data, alpha=0.05):

    # Step 1: Mengambil nilai dari tabel 2x2
    a = data[0][0]
    b = data[0][1]
    c = data[1][0]
    d = data[1][1]

    # Step 2: Menghitung total baris
    baris1 = a + b
    baris2 = c + d

    # Step 3: Menghitung total kolom
    kolom1 = a + c
    kolom2 = b + d

    # Step 4: Menghitung grand total
    N = a + b + c + d

    # Step 5: Menghitung probabilitas Fisher
    p_value = (
        factorial(baris1)
        * factorial(baris2)
        * factorial(kolom1)
        * factorial(kolom2)
    ) / (
        factorial(a)
        * factorial(b)
        * factorial(c)
        * factorial(d)
        * factorial(N)
    )

    # Step 6: Mengambil keputusan
    if p_value < alpha:
        keputusan = "Tolak H0: terdapat hubungan"
    else:
        keputusan = "Gagal menolak H0: tidak terdapat hubungan"

    # Step 7: Menampilkan hasil
    print("a =", a)
    print("b =", b)
    print("c =", c)
    print("d =", d)
    print("baris1 =", baris1)
    print("baris2 =", baris2)
    print("kolom1 =", kolom1)
    print("kolom2 =", kolom2)
    print("grand_total =", N)
    print("p_value =", p_value)
    print("keputusan =", keputusan)

data = [
    [30, 20],
    [15, 35]]
fisher_exact_test(data)
```
